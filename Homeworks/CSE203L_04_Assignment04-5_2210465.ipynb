```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left_child = None
        self.right_child = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def inorder(self, node):
        if node is None:
            return
        else:
            self.inorder(node.left_child)
            print(node.data)
            self.inorder(node.right_child)

    #2
    def insertR(self, node, value):
        if node is None:
            node = Node(value)
            return node
        else:
            if value <= node.data:
                node.left_child  = self.insertR(node.left_child, value)
            else:
                node.right_child  = self.insertR(node.right_child , value)
            return node
    #4
    def find_minR(self, node):
            if node == None:
                return None
            elif node.left_child == None:
                return node.data
            else:
                return self.find_minR(node.left_child)

    #6
    def check_leaf_nodes(self, node):
        if node is None:
            return []

        if node.left_child is None and node.right_child is None:
            if node.data % 2 == 0:
                return ['even']
            else:
                return ['odd']
        
        left_result = self.check_leaf_nodes(node.left_child)
        right_result = self.check_leaf_nodes(node.right_child)

        return left_result + right_result

#8
def countNodes(root):
    if root is None:
        return 0
    else:
        return 1 + countNodes(root.left_child) + countNodes(root.right_child)

#10
def countLeafNodes(root):
    if root is None:
        return 0
    elif root.left_child is None and root.right_child is None:
        return 1
    else:
        return countLeafNodes(root.left_child) + countLeafNodes(root.right_child)

    
```

```python
bst = BinarySearchTree()

values = [7, 15, 9, 8, 10, 5, 3, 16, 11, 17, 13]
for value in values:
    bst.root = bst.insertR(bst.root, value)

bst.inorder(bst.root)
print()

print('Min:',bst.find_minR(bst.root))
print()

print(bst.check_leaf_nodes(bst.root))
print()

print("Number of nodes:",countNodes(bst.root))
print()

print("Number of leaf nodes:",countLeafNodes(bst.root))

```
