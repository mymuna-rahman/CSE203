1. Write a recursive program to find the sum of first n numbers.

sum = 1 +2 + 3 + …….n

```python
def sumR(n):
    if n==1:
        return 1
    else:
        return n+sumR(n-1)

print(sumR(3))
print(sumR(5))
```

2. Given a string, check whether the string is a palindrome or not . Write the recursive function. The function will return true or false

```python
def isPalindromeR(string):

    if len(string)==0:
        return True
    else:
        return string[0]==string[-1] and isPalindromeR(string[1:-1])

print(isPalindromeR('madam'))
print(isPalindromeR('madamm'))
print(isPalindromeR(''))
```

3. Given a integer number, check whether it is palindrome or not. Write recursive function. It will return true or false

```python
def reverseR(num,rev=0): #this function is to just reverse the number
    if num==0:
        return rev
    else:
        digit=num%10
        rev=rev*10+digit
        return reverseR(num//10,rev)

print(reverseR(123)) #the main function where it checks if the number equals the reverse from the previous function

def isPalindromeR_int(num):
    if num!=reverseR(num):
        return False
    else:
        return True

print(isPalindromeR_int(1221))
```

4. Given two integers, find and print the GCD (Greatest Common Divisor) of them. The function will return the GCD

```python
def gcdR(n1, n2, div=None):
    if div is None:  
        div = min(n1, n2) 

    if n1 % div == 0 and n2 % div == 0:  
        return div
    else:
        return gcdR(n1, n2, div - 1)  

print(gcdR(12, 8))
print(gcdR(7, 13))
```

5. Reverse a string using Recursion

```python
def revStringR(s, temp=''):
    if len(s)==0:
        return temp
    else:
        temp+=s[-1]
        return temp + revStringR(s[:-1])

print(revStringR("Python"))
print(revStringR("HeLLo WoRlD"))
print(revStringR("1363"))
```

8. This problem is about generating Power set in lexicographical order. Examples :

• Input : abc

• Output : a ab abc ac b bc c

```python
def powerset(s):
    if len(s) == 0:
        return ['']

    first = s[0]
    rest = s[1:]
    rest_powerset = powerset(rest)
    rest_powerset_with_first = [first + subset for subset in rest_powerset]
    return rest_powerset + rest_powerset_with_first


def generate_power_set(s):
    powerset_list = powerset(s)
    powerset_list.sort()
    powerset_list.remove('')
    return powerset_list


input_string = "abc"
output = generate_power_set(input_string)
print(output)
```

10. Given a doubly linked list. Reverse it using recursion. Here we assume, we keep self.head pointer. We are not keeping self.tail.

Given Linked List:



```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            cur = self.head
            while cur.next:
                cur = cur.next
            cur.next = new_node
            new_node.prev = cur

    def print_list(self):
        cur = self.head
        while cur:
            print(cur.data)
            cur = cur.next

    def reverse_recursive(self, node):
        if node is None:
            return None
        if node.next is None:
            self.head = node
            return node
        reversed_tail = self.reverse_recursive(node.next)
        temp = node.next
        temp.prev = None
        temp.next = node
        node.prev = temp
        node.next = None

        return reversed_tail

    def reverse(self):
        self.head = self.reverse_recursive(self.head)


dll = DoublyLinkedList()
dll.append(1)
dll.append(2)
dll.append(3)
dll.append(4)
dll.append(5)
print("Original List:")
dll.print_list()

dll.reverse()
print("Reversed List:")
dll.print_list()
```

11. Write a recursive function minRec(arr, n) to find the minimum element of an array. arr is the array and n is the length of the arr.

```python
def minRec(arr, n):
    if n == 1:
        return arr[0]
    else:
        min_of_rest = minRec(arr, n-1)
        return arr[0] if arr[0] < min_of_rest else min_of_rest

arr = [5, 2, 1, 9, 4]
n = len(arr)
print(minRec(arr, n)) 
```
