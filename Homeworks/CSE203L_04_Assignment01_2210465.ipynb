```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None

class SLL:
    def __init__(self):
        self.head=None
        self.size=0

    def append(self,data):
        if self.head==None:
            newnode=node(data)
            self.head=newnode
        else:
            current=self.head
            while current.next is not None:
                current=current.next
            newnode=node(data)
            current.next=newnode
        self.size+=1

    def printlist(self):
        if self.head==None:
            print("List is empty.")
        else:
            current = self.head
            while current != None:
                print(current.data, end ='->')
                current = current.next
        print()   
          
    #3
    def find_intersection(list1, list2):
        len1=0 
        len2=0
        curr1=list1.head
        curr2=list2.head

        while curr1:
            len1+=1
            curr1=curr1.next

        while curr2:
            len2+=1
            curr2=curr2.next

        if len1<len2:
            list1=list2
            list2=list1

        curr1=list1.head
        curr2=list2.head

        for _ in range(len1-len2):
            curr1=curr1.next

        while curr1!=curr2:
            curr1=curr1.next
            curr2=curr2.next

        return curr1

    #4
    def remove_node(self,n):
        prev = None
        current = self.head

        while current!=None:
            if current.data == n:
                if prev == None:
                    self.head = current.next
                else:
                    prev.next = current.next
                current = current.next
            else:
                prev = current
                current = current.next

    #5
    def reverse_sll(self):
        if self.head is None:
            print('List is empty.')
            return

        prev = None
        current = self.head

        while current:
            next = current.next
            current.next = prev
            prev = current
            current = next

        self.head = prev


sll1 = SLL()
sll1.append(1)
sll1.append(2)
sll1.append(3)
sll1.append(4)
sll1.append(5)

sll2 = SLL()
sll2.append(6)
sll2.append(7)
sll2.append(8)
sll2.append(3)
sll2.append(4)
sll2.append(5)
sll1.find_intersection(sll2)



print("Remove all nodes of a given value:")
s=SLL() 
s.append(1)
s.append(2)
s.append(3)
s.append(2)
s.append(4)
s.append(4)
s.append(5)
s.append(2)
s.printlist()
s.remove_node(2)
s.printlist()       
print()
print('Reverse SLL:') 
s1=SLL()
s1.append(1)
s1.append(2)
s1.append(3)
s1.append(4)
s1.printlist()
s1.reverse_sll()
s1.printlist()
```

After Insertion (value 4 at position 2):
Output: 1 -> 4 -> 2 -> 3

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.prev=None

class DLL:
    def __init__(self):
        self.head=None
        self.tail=None

    def append(self,data):
        newnode=node(data)
        
        if self.head==None:
            self.head=newnode
            self.tail=newnode
        else:
            current=self.head
            while current.next is not None:
                current=current.next
            current.next=newnode
            newnode.prev=current
            self.tail=newnode

    def printlist(self):
        current=self.head
        print(' <- ',end="")
        while current!=None:
            if current.next==None:
                print(current.data, end=" -> ")
            else:
                print(current.data, end=" <-> ")
            current=current.next
        print()

    #1
    def length(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count
    def rotate_right(self, k):
        length = self.length()
        if length <= 1 or k == 0:
            return
            
        if k == 0:
            return

        current = self.head

        for _ in range(length - k - 1):
            current = current.next

        new_tail = current
        new_head = current.next

        while current.next:
            current = current.next

        current.next = self.head
        self.head.prev = current

        new_tail.next = None
        self.head = new_head
        new_head.prev = None
  
    #2
    def nth_from_end(self, n):
        if self.head is None:
            print("List is empty.")
            return None

        current = self.head
        nth_from_end = self.head

        for i in range(n):
            if current is None:
                print("The position is out of range.")
                return None
            current = current.next

        while current is not None:
            current = current.next
            nth_from_end = nth_from_end.next

        return nth_from_end.data

    #4
    def reverse_dll(self):
        current=self.tail
        print(' <- ',end="")
        while current!=None:
            if current.prev == None:
                print(current.data, end=" -> ")
            else:
                print(current.data, end=" <-> ")
            current=current.prev
        print()

    #5
    def insert_at_position(self, n, p):
        newnode=node(n)

        if self.head is None:
            if p == 0:
                self.head = newnode
            return
        if p==1:
            newnode.next=self.head
            self.head.prev=newnode
            self.head=newnode
        else:
            current=self.head
            curr_pos=0

            while current and curr_pos<p:
                prev=current
                current=current.next
                curr_pos+=1
            if curr_pos==p:
                prev.next=newnode
                newnode.prev=prev
                newnode.next=current
            else:
                print("The position is out of range.")

print("Reverse DLL:")
d1 = DLL()
d1.append(1)
d1.append(2)
d1.append(3)
d1.append(4)
d1.append(5)
d1.reverse_dll()
d1.printlist()
print()

print("Insert at specific position:")
d1.insert_at_position(13,3)
d1.printlist()
d1.insert_at_position(20,10)
d1.insert_at_position(0,1)
d1.printlist()

d2=DLL()
d2.insert_at_position(3,0)
d2.printlist()
print()

print("Nth node from end:")
d3=DLL()
d3.append(1)
d3.append(2)
d3.append(3)
d3.append(4)
d3.printlist()
print(d3.nth_from_end(2))
print()

print("Rotate right")
d4=DLL()
d4.append(1)
d4.append(2)
d4.append(3)
d4.append(4)
d4.append(5)
d4.append(6)
d4.printlist()
d4.rotate_right(2)
d4.printlist()



```

5. Write a function to check if a queue is palindrome or not. The function will return True or False
Example 1:
queue1: [1,2,3,4,5]
Output: False
Example 2:
queue2: [1,2,3,2,1]
Output: True

```python
class Stack:
    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size()==0

    def push(self,value):
        self.data.append(value)

    def pop(self):
        if self.isEmpty():
            print('Stack is empty')
        else:
            self.data.pop()

    def peek(self):
        if self.isEmpty():
            print('Stack is empty')
        else:
            return self.data[-1]

    def printStack(self):
        if self.isEmpty():
            print('Stack is empty')
        else:
            print(self.data)

    #1
    def is_palindrome(self):
        size=len(self.data)
        if size<=1:
            return True
        
        temp=list(self.data)
        
        while len(temp)>1:
            if temp.pop(0) != temp.pop():
                return False
        return True

print("Is stack palindrome:")
stack1=Stack()
stack1.push(1)
stack1.push(2)
stack1.push(3)
stack1.push(4)
stack1.push(5)
stack1.printStack()
print(stack1.is_palindrome()) 

stack2=Stack()
stack2.push(1)
stack2.push(2)
stack2.push(3)
stack2.push(2)
stack2.push(1)
stack2.printStack()
print(stack2.is_palindrome()) 
```

```python
#2
def remove_largest(stack):
    helpingstack=Stack()
    max=stack.peek()
    while stack.isEmpty() is not True:
        val=stack.peek()
        if val>max:
            max=val
        helpingstack.push(val)
        stack.pop()

    while helpingstack.isEmpty() is not True:
        val=helpingstack.peek()
        if val is not max:
            stack.push(val)
        helpingstack.pop()
    return stack

s=Stack()
for i in (1,10,9,7,2):
    s.push(i)
s.printStack()
remove_largest(s)
s.printStack()
```

```python
#3
def balanced(string):
    s=Stack()
    open_brackets=['(','{','[']
    close_brackets=[')','}',']']
    for brac in string:
        if brac in open_brackets:
            s.push(brac) 
        elif brac in close_brackets:
            index = close_brackets.index(brac) 
            if s.peek()==open_brackets[index]: 
                s.pop() 
            else:
                return False     
        else:
            pass 
    if s.isEmpty()==True: 
        return True
    else:
        return False
            
input1='a(bc{]})'
print(input1)
if balanced(input1):
    print('True')
else:
    print('False')

input2='({[[]]})'
print(input2)
if balanced(input2):
    print('True')
else:
    print('False')
```
